```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 불러오기
data = pd.read_csv("nightlife_crime.csv", encoding='cp949')

# 'murder'부터 'violence_Tot'까지의 열을 합산하여 'total_crime' 열 생성
data['total_crime'] = data.loc[:, 'murder':'violence_Tot'].sum(axis=1)

# 'region', 'tot_nightlife', 'total_crime' 열 선택
subset_data = data[['region', 'tot_nightlife', 'total_crime']]

# 'region'을 기준으로 그룹화하고 합계 계산
grouped_data = subset_data.groupby('region').sum().reset_index()

# 시각화
plt.figure(figsize=(12, 6))

# tot_nightlife 값 시각화
sns.barplot(x='region', y='tot_nightlife', data=grouped_data, color='skyblue', label='tot_nightlife')

# total_crime 값 시각화 (두 번째 y축 사용)
ax2 = plt.twinx()
sns.barplot(x='region', y='total_crime', data=grouped_data, color='orange', label='total_crime', ax=ax2)

# 축과 레이블 설정
plt.ylabel('tot_nightlife', color='blue')
ax2.set_ylabel('total_crime', color='red')
plt.title('Region별 tot_nightlife 및 총 범죄 발생량')

plt.show()
```

```python
# 가설 수립 : 환경이 청소년 범죄발생에 영향을 미침

# 유흥업소에 포함되는 변수 : Pub, Karaoke
# 전체 비율 시각화
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 새로운 데이터 파일 불러오기
extracted_data = pd.read_csv("extracted_data_.csv", encoding='cp949')

# 'tot_nightlife'와 'total_teen_crime'의 상관 계수 계산
correlation = extracted_data['tot_nightlife'].corr(extracted_data['total_teen_crime'])

# 시각화
plt.figure(figsize=(8, 6))
sns.scatterplot(x='tot_nightlife', y='total_teen_crime', data=extracted_data, alpha=0.7)
plt.title(f'상관 계수: {correlation:.2f}')
plt.xlabel('tot_nightlife')
plt.ylabel('total_teen_crime')
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm

# 새로운 데이터 파일 불러오기
extracted_data = pd.read_csv("extracted_data_.csv", encoding='cp949')

# 독립 변수(X)와 종속 변수(y) 설정
X = extracted_data[['tot_nightlife']]
y = extracted_data['total_teen_crime']

# 상수 항 추가
X = sm.add_constant(X)

# 단순 선형 회귀 모델 피팅
model = sm.OLS(y, X).fit()

# 회귀 모델 요약 확인
print(model.summary())
```

```python
# 회귀분석

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(extracted_data[["tot_nightlife"]], extracted_data[["total_teen_crime"]])
print(model.coef_) # 회귀계수
print(model.intercept_) # 절편

# 결정계수 구하기
model.score(extracted_data[["tot_nightlife"]], extracted_data[["total_teen_crime"]])
```

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

# 산점도와 회귀선 시각화
plt.figure(figsize=(8, 6))
sns.scatterplot(x='tot_nightlife', y='total_teen_crime', data=extracted_data, alpha=0.7, label='실제 데이터')
plt.plot(X['tot_nightlife'], model.predict(X), color='red', label='회귀선')
plt.title(f'상관 계수: {correlation:.2f}')
plt.xlabel('tot_nightlife')
plt.ylabel('total_teen_crime')
plt.legend()
plt.show()
```
