```python
# 1. 여성 생존비율, 남성 생존 비율 비교 시각화
colors = ['green', 'gold']

data = titanic["생존여부"].groupby(titanic['성별'])
plt.pie(data.size(), labels=['남자', '여자'], autopct='%.1f%%', startangle=260, counterclock=False,
        shadow=True, colors=colors)
plt.title("여성 생존비율, 남성 생존 비율 비교 시각화")
plt.show()
```

```python
# 2. 객실 등급별 생존비율 비교 시각화
colors = ['gold', 'silver','green']

data = titanic["생존여부"].groupby(titanic['객실등급'])
plt.pie(data.size(), labels=['1', '2', '3'], autopct='%.1f%%', startangle=260, counterclock=False,
        shadow=True, colors=colors)
plt.title("객실 등급별 생존비율 비교 시각화")
plt.show()
```

```python
# 3. 승선항구(Embarked ) 별 생존비율 비교 시각화
colors = ['gold', 'silver','green']

data = titanic["생존여부"].groupby(titanic['Embarked'])
plt.pie(data.size(), labels=['S', 'C', 'Q'], autopct='%.1f%%', startangle=260, counterclock=False,
        shadow=True, colors=colors)
plt.title("승선항구(Embarked ) 별 생존비율 비교 시각화")
plt.show()
```

```python
# 4.  생존한 승객과 그렇지 않은 승객의 연령이 통계학적으로 차이가 난다고 할 수 있는지 
# 95% 신뢰수준으로t 검정하세요.
import numpy as np
from scipy import stats
```

```python
df_new = titanic[['생존여부', '나이']]
df_new
```

```python
pd.isna(df_new)
```

```python
# 결측치 제거
df_new = df_new.dropna(how ='any')  #how='all' 모든 속성이 결측치일때만 해당 행 삭제 
df_new
```

```python
# 일표본 t-test
# 귀무가설: 두 집단의 평균의 나이차가 0이다.
# 대립가설: 두 집단의 평균의 나이차가 0이 아니다.

alive = df_new.loc[ (df_new.생존여부 == 1), ("나이")]
dead = df_new.loc[ (df_new.생존여부 == 0), ("나이")]

print(alive.count())
print(dead.count())

tTestResult = stats.ttest_ind(alive, dead)
print(tTestResult)
```

```python
# 검정 결과 p-value는 0.0391이다.
# p-value가 0.05보다 작으므로 대립가설은 채택하고, 귀무가설이 기각된다.
```
