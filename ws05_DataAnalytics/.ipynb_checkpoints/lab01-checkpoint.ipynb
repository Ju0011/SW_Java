```python
# 문제 9) 전체 붓꽃 품종별 꽃받침폭 분포 히스토그램

plt.rcParams['axes.unicode_minus'] = False

print(iris.Name.unique())

plt.hist(iris.loc[(iris.Name == "Iris-versicolor"), ["SepalWidth"] ],
      bins=20, alpha = 0.2, histtype='stepfilled', label="Versicolor")
plt.hist(iris.loc[(iris.Name == "Iris-setosa"), ["SepalWidth"] ],
      bins=20, alpha = 0.2, histtype='stepfilled', color="red", label="Setosa")
plt.hist(iris.loc[(iris.Name == "Iris-virginica"), ["SepalWidth"] ],
      bins=20, alpha = 0.2, histtype='stepfilled', label="Virginica")
plt.ylabel("빈도") ;   plt.xlabel("꽃받침폭")
plt.legend(prop={'size': 12}, title = '붓꽃 품종')
plt.title ("붓꽃 품종별 꽃받침폭 분포 ")
```
