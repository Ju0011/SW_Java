```python
import pandas as pd
from konlpy.tag import Okt

# CSV 파일에서 데이터프레임으로 읽어오기
df = pd.read_csv('baseball_crawl.csv', encoding='utf-8') #cp949

okt = Okt()
all_nouns = []  # 모든 nouns를 저장할 리스트

# Title과 Article 내용을 합쳐서 하나의 문자열로 만들기
combined_content = df['Title'] + ' ' + df['Article']

# 하나의 문자열로 합쳐진 내용을 형태소 분석하기
for content in combined_content:
    nouns = okt.nouns(str(content))
    all_nouns.extend(nouns)  # 모든 nouns를 하나의 리스트에 추가

print(all_nouns)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# 'cleaned_bb.txt' 파일에서 단어들을 읽어옴
with open('cleaned_bb.txt', 'r', encoding='utf-8') as f:
    samples = f.readlines()

# 개행 문자 제거 및 소문자로 변환하여 단어 리스트 생성
samples = [line.strip().lower() for line in samples]

# Tokenizer를 사용하여 단어를 인덱스로 변환
tokenizer = Tokenizer()
tokenizer.fit_on_texts(samples)

# 텍스트 데이터를 숫자 시퀀스로 변환
sequences = tokenizer.texts_to_sequences(samples)

# 시퀀스를 패딩하여 길이를 맞춤
maxlen = 100  # 시퀀스의 최대 길이
data = pad_sequences(sequences, maxlen=maxlen)

# 레이블 생성 (여기서는 임의로 생성)
labels = np.random.randint(2, size=len(samples))  # 이진 분류를 가정

# 데이터를 훈련 세트와 테스트 세트로 분할
train_size = int(len(data) * 0.8)
train_data = data[:train_size]
train_labels = labels[:train_size]
test_data = data[train_size:]
test_labels = labels[train_size:]

print(samples[0])  # 첫 번째 샘플 텍스트 출력
print(train_data[0][train_data[0] != 0])  # 첫 번째 패딩되지 않은 시퀀스 요소 출력


# 이제 train_data, train_labels, test_data, test_labels를 사용하여 모델을 학습하고 평가할 수 있습니다.
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# 'cleaned_bb.txt' 파일에서 단어들을 읽어옴
with open('clean_esport.txt', 'r', encoding='cp949') as f:
    samples = f.readlines()

# 개행 문자 제거 및 소문자로 변환하여 단어 리스트 생성
samples = [line.strip().lower() for line in samples]

# Tokenizer를 사용하여 단어를 인덱스로 변환
tokenizer = Tokenizer()
tokenizer.fit_on_texts(samples)

# 텍스트 데이터를 숫자 시퀀스로 변환
sequences = tokenizer.texts_to_sequences(samples)

# 시퀀스를 패딩하여 길이를 맞춤
maxlen = 100  # 시퀀스의 최대 길이
data = pad_sequences(sequences, maxlen=maxlen)

# 레이블 생성 (여기서는 임의로 생성)
labels = np.random.randint(2, size=len(samples))  # 이진 분류를 가정

# 데이터를 훈련 세트와 테스트 세트로 분할
train_size = int(len(data) * 0.8)
train_data = data[:train_size]
train_labels = labels[:train_size]
test_data = data[train_size:]
test_labels = labels[train_size:]

print(samples[0])  # 첫 번째 샘플 텍스트 출력
print(train_data[0][train_data[0] != 0])  # 첫 번째 패딩되지 않은 시퀀스 요소 출력


# 이제 train_data, train_labels, test_data, test_labels를 사용하여 모델을 학습하고 평가할 수 있습니다.
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# 'cleaned_bb.txt' 파일에서 단어들을 읽어옴
with open('clean_esport.txt', 'r', encoding='cp949') as f:
    samples = f.readlines()

# 개행 문자 제거 및 소문자로 변환하여 단어 리스트 생성
samples = [line.strip().lower() for line in samples]

# Tokenizer를 사용하여 단어를 인덱스로 변환
tokenizer = Tokenizer()
tokenizer.fit_on_texts(samples)

# 텍스트 데이터를 숫자 시퀀스로 변환
sequences_2 = tokenizer.texts_to_sequences(samples)

sequences_2
```

```python
len(sequences_2)
```

```python
#### 데이터 준비 - 원핫인코딩
import numpy as np

def vectorize_sequences(sequences, dimension=10000):
    # 크기가 (len(sequences), dimension))이고 모든 원소가 0인 행렬을 만듭니다
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.  # results[i]에서 특정 인덱스의 위치를 1로 만듭니다
    return results
    
# 훈련 데이터를 벡터로 변환합니다
x_train = vectorize_sequences(sequences_2)
# 테스트 데이터를 벡터로 변환합니다
#x_test = vectorize_sequences(sequences)
```

```python
x2_df = pd.DataFrame(x_train)
x2_df
```

```python
x2_df['label'] = int(1)
x2_df
```

```python
total = pd.concat([x_df,x2_df])
len(total)
```

```python
train_labels = total['label']
train_labels
```

```python
# 레이블을 벡터로 바꿉니다
y_train = np.asarray(train_labels).astype('float32')
#y_test = np.asarray(test_labels).astype('float32') 

len(y_train)
```

```python
# 신경망 모델 만들기
# 케라스 구현

from tensorflow.keras import models
from tensorflow.keras import layers

model = models.Sequential() 
model.add(layers.Dense(16, activation = 'relu', input_shape =(10000,))) 
model.add(layers.Dense(16, activation = 'relu')) 
model.add(layers.Dense(1, activation = 'sigmoid'))

model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
x_val = total[:24555]
partial_x_train = total[24555:]

y_val = y_train[:24555]
partial_y_train = y_train[24555:]

partial_x_train
```

```python
history = model.fit(partial_x_train,
                   partial_y_train,
                   epochs = 20,
                   batch_size = 512,
                   validation_data = (x_val, y_val)) 
```
