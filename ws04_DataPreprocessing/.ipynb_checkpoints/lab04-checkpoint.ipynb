```python
import pandas as pd 
# titanic.csv 를 데이터프레임으로 read  , encoding = "euc-kr"  
data = pd.read_csv("titanic.csv",encoding = "euc-kr")
data
```

```python
# 앞의 7개 행만 출력 
data.head(7)
```

```python
# 생존한 승객 중 여성의 나이와 객실 등급(pclass) 을 출력 
data.loc[ (data.생존여부 == 1)&(data.성별 == 'female'),("나이", "객실등급")]
```

```python
#  생존한 승객 중 여성 수
data.loc[ (data.생존여부 == 1)&(data.성별 == 'female'),("PassengerId")].count()
```

```python
#  생존한 승객 중 남성 수 
data.loc[ (data.생존여부 == 1)&(data.성별 == 'male'),("PassengerId")].count()
```

```python
# 전체 남성 승객 수
data.loc[(data.성별 == 'male'),("PassengerId")].count()
```

```python
#  생존한 승객 중 여성이면서 1등칸 객실 승객 수 
data.loc[ (data.생존여부 == 1)&(data.성별 == 'female')&(data.객실등급 == 1), ("PassengerId")].count()
```

```python
#  생존한 승객 중 여성이면서 3등칸 객실 승객 수 
data.loc[ (data.생존여부 == 1)&(data.성별 == 'female')&(data.객실등급 == 3), ("PassengerId")].count()
```

```python
# 1등칸 전체 여성승객 수 
data.loc[ (data.성별 == 'female')&(data.객실등급 == 1),("PassengerId")].count()
```

```python
# 3등칸 전체 여성승객 수 
data.loc[ (data.성별 == 'female')&(data.객실등급 == 3), ("PassengerId")].count()
```

```python
# 전체 승객 수
data["PassengerId"].count()
```

```python
# 결측치 갯수 
pd.isna(data)
#pd.isna(data).count()
```

```python
# 13세 미만 전체 승객 수 
data.loc[ (data.나이 < 13), ("PassengerId")].count()
```

```python
# 생존한 13세 미만 전체 승객 수
data.loc[ (data.생존여부 == 1)&(data.나이 < 13), ("PassengerId")].count()
```
