```python
import pandas as pd
data = pd.read_csv("lab02_subway2013.csv")
print(data)
```

```python
# 1) 위의 파일을 읽고, 5행만 출력 
print(data[4:5])
```

```python
# 2) 행 제목만 출력
print(data.index)
```

```python
# 3) 열 제목만 출력
print(data.columns)
```

```python
# 4)“평일” 환승인원을 기준으로 오름차순 정렬하여 출력
print(data.sort_values(by='평일', ascending=True ))
```

```python
# 5) 토요일 환승 인원이 100000명 이상의 역의 역명과 다른 요일 환승인원 출력
data[ (data["토요일"]>  100000) ]
```

```python
#6) 일요일 환승 인원만 역명과 함께 출력
print(data[['역명','일요일']])
```

```python
# 7) 11~20번째 행의 일요일 환승 인원만 역명과 함께 출력
# new_data = (data[['역명','일요일']])
# print(new_data[11:21])

print(data[['역명','일요일']].iloc[11:21])
```
