```python
import pandas as pd
data = pd.read_excel("지역_위치별(주유소).xls")
print(data)
```

```python
# 1) 결측치가 존재하는 행 제거
# df 복제 
# nan인 값에만 True가 표시되게 하는 함수
df_new = data.copy() 
pd.isna(df_new)
```

```python
#결측치를 가지고 있는 행들을 삭제   #how='any'하나라도 결측치가 있으면 해당 행 삭제
df_new = df_new.dropna(how ='any')  #how='all' 모든 속성이 결측치일때만 해당 행 삭제 
df_new
```

```python
# 2) 이상치가 존재하는 지 확인하고, 이상치가 존재하는 행 제거 
# Basic box plot
import matplotlib.pyplot as plt
%matplotlib inline


a =plt.boxplot(df_new['고급휘발유'])
plt.show()
```
