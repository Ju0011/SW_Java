```python
import mglearn
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc

import warnings
warnings.filterwarnings('ignore')

#한글 폰트 등록
font_location = "c:/Windows/fonts/malgun.ttf"   
font_name = font_manager.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)

iris = pd.read_csv('iris.csv')
iris.head()
```

### 클러스터링 

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(iris.loc[:,["SepalLength", "SepalWidth","PetalLength","PetalWidth"]])
```

### PCA로 차원 축소

```python
## 표준화 

from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler() 
X_scaled = scaler.fit_transform (iris.loc[:,["SepalLength", "SepalWidth","PetalLength","PetalWidth"]])
```

```python
## 주성분 분석
from sklearn.decomposition import PCA 
pca = PCA(n_components=2) 
# pca.fit( iris.loc[:,["SepalLength", "SepalWidth","PetalLength","PetalWidth"]] ) 
#iris_pca = pca.transform ( iris.loc[:,["SepalLength", "SepalWidth","PetalLength","PetalWidth"]] )

iris_pca =pca.fit_transform(  X_scaled ) 
```

### PCA 후 주성분으로 시각화

```python
mglearn.discrete_scatter(iris_pca[:,0] , iris_pca[:,1], kmeans.labels_  ) 
plt.legend(["1", "2", "3"], loc = 'best')
plt.xlabel("주성분1")
plt.ylabel("주성분2")
plt.title( "클러스터링 결과")
plt.show() 

mglearn.discrete_scatter(iris_pca[:,0] , iris_pca[:,1], iris["Name"]   ) 
plt.legend(["1", "2", "3"], loc = 'best')
plt.xlabel("주성분1")
plt.ylabel("주성분2")
plt.title( "실제 품종별 데이터")
plt.show() 
```

```python
mglearn.discrete_scatter(iris["PetalLength"], iris["PetalWidth"], kmeans.labels_  ) 
plt.legend(["1", "2", "3"], loc = 'best')
plt.xlabel("PetalLength")
plt.ylabel("PetalWidth")
plt.title( "클러스터링 결과")
plt.show() 

mglearn.discrete_scatter(iris["PetalLength"], iris["PetalWidth"], iris["Name"]   )
plt.legend(["1", "2", "3"], loc = 'best')
plt.xlabel("PetalLength")
plt.ylabel("PetalWidth")
plt.title( "실제 품종별 데이터")
plt.show()
```
