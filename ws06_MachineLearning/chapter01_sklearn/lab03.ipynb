```python
import mglearn
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc

import warnings
warnings.filterwarnings('ignore')

#한글 폰트 등록
font_location = "c:/Windows/fonts/malgun.ttf"  #(windows)
font_name = font_manager.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)

iris = pd.read_csv('iris.csv')
iris.head()
```

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(iris.loc[:,["SepalLength", "SepalWidth","PetalLength","PetalWidth"]])
```

```python
mglearn.discrete_scatter(iris["SepalLength"], iris["SepalWidth"], kmeans.labels_  ) 
plt.legend(["1", "2", "3"], loc = 'best')
plt.xlabel("SepalLength")
plt.ylabel("SepalWidth")
plt.show() 

mglearn.discrete_scatter(iris["SepalLength"], iris["SepalWidth"], iris["Name"]   )
plt.legend(["1", "2", "3"], loc = 'best')
plt.xlabel("SepalLength")
plt.ylabel("SepalWidth")
plt.show()
```
